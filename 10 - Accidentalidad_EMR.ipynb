Importamos las librerías necesarias

```pyspark
import sys
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

Obtenemos la sesion de spark

```pyspark
spark = SparkSession\
    .builder\
    .appName("SparkETL")\
    .getOrCreate()
```

Cargamos los datos desde **AWS S3**. 

Nota: Cambie la ruta para que coincida con el bucket creado

```pyspark
data= spark.read.option("inferSchema", "true").option("header", "true").csv("s3://asd123qwe/input/accidentalidad.csv")
```
